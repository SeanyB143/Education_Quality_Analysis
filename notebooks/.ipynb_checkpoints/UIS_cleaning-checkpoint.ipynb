```python
import pandas as pd
import numpy as np
```

```python
uis_1 = pd.read_csv('../data/UIS/EDULIT_DS_23082019165936814.csv')
uis_2 = pd.read_csv('../data/UIS/EDULIT_DS_25082019015237837.csv')

uis = pd.concat([uis_1, uis_2], sort=True)
uis.head()
```

```python
uis_df = pd.DataFrame(index=features)

for year in years:
    for country in countries:
        year_country_dict = {feature: 0 for feature in features}
        for feature in features:
            country_filter = (uis['Country'] == country)
            year_filter = (uis['TIME'] == year)
            feature_filter = (uis['Indicator'] == feature)
            try:
                year_country_dict[feature] += uis[country_filter & year_filter
                                                  & feature_filter]['Value'][0]
            except:
                year_country_dict[feature] = np.nan
                
        uis[f"{year}_{country}"] = pd.Series(year_country_dict)
```

```python
def create_df(data, name=None, countries=countries):
    years = list(data['TIME'].unique())

    index_mean = [f"{year}_mean_{name}" for year in years]
#     index_std = [f"{year}_std_{name}" for year in years]

#     output_df = pd.DataFrame(index=index_mean + index_std, columns=countries)
    output_df = pd.DataFrame(index=index_mean, columns=countries)
    
    for year in years:
        for country in countries:
            mean = str(data[(data['Year'] == year) & (data['Jurisdiction'] == country)]['Average'].values[0])
#             std = str(data[(data['Year'] == year)
#                            & (data['Jurisdiction'] == country)]['Standard Error'].values[0]).strip('()')
            try:
                output_df[country][f"{year}_mean_{name}"] = float(mean)
            except:
                output_df[country][f"{year}_mean_{name}"] = np.nan
#             try:
#                 output_df[country][f"{year}_std_{name}"] = float(std)
#             except:
#                 output_df[country][f"{year}_std_{name}"] = np.nan
    return output_df
```
